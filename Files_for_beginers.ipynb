FOR IMAGE AND AUDIO FILES

```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("sample_image.png")
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
import wave
import numpy as np
import matplotlib.pyplot as plt

with wave.open("sample_audio.wav", "rb") as wf:
    n_frames = wf.getnframes()
    audio_data = np.frombuffer(wf.readframes(n_frames), dtype=np.int16)

plt.plot(audio_data[:500])  # Show waveform (first 500 samples)
plt.title("Audio Waveform")
plt.show()
```
